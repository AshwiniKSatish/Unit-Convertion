```python
import tkinter as tk
from tkinter import ttk

class UnitConverterApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Unit Converter")

        self.create_widgets()

    def create_widgets(self):
        
        label_type = ttk.Label(self.root, text="Conversion Type:")
        label_type.grid(row=0, column=0)
        types = ["Centimeter to Feet", "Feet to Inches", "Sqft to Sqmtrs", "Sqft to Acres/Hectres"]
        self.optiontype = ttk.Combobox(self.root, values=types)
        self.optiontype.grid(row=0, column=1)

        label_value = ttk.Label(self.root, text="Enter Value:")
        label_value.grid(row=2, column=0, padx=10, pady=10)

        self.entry_value = ttk.Entry(self.root)
        self.entry_value.grid(row=2, column=1, padx=10, pady=10)

        convert_button = ttk.Button(self.root, text="Convert", command=self.convert_length)
        convert_button.grid(row=3, padx=10, pady=10)

        self.result_label = ttk.Label(self.root, text="Result: ")
        self.result_label.grid(row=4, padx=10, pady=10)

    def convert_length(self):
        value = float(self.entry_value.get())
        conv_type = self.optiontype.get()
        result_text = ""

       
        if conv_type == "Centimeter to Feet":
            result_text = "{} cm is equal to {} feet.".format(value, value / 30.48)
        elif conv_type == "Feet to Inches":
            result_text = "{} feet is equal to {} inches.".format(value, value * 12)
        elif conv_type == "Sqft to Sqmtrs":
            result_text = "{} sqft is equal to {} sqmtrs".format(value, value * 0.0929)
        elif conv_type == "Sqft to Acres/Hectres":
            result_text = "{} sqft is equal to {} acres and {} sqft is equal to {} hectre".format(value,value / 43560,value,value / 107639.104)

        self.result_label.config(text=result_text)


if __name__ == "__main__":
    root = tk.Tk()
    app = UnitConverterApp(root)
    root.mainloop()

    def update_conversion_types(event):
        selected_type = combo_type.get()
        combo_conversion_type['values'] = conversion_type_options.get(selected_type, [])

    combo_type.bind("<<ComboboxSelected>>", update_conversion_types)

    root.mainloop()
```
